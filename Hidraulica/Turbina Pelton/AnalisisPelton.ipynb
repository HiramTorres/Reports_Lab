## Datos para muestra a una presión de 1.2 bares

El caudal fue se determinó para un $V(m^3)  = 0.0459885 $ en un tiempo de $t = 51 s$ , $Q = 0.000901735$

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme(style = "darkgrid")
id = pd.Series(range(1,7), name="id")
db = pd.read_excel("Practica turbina Pelton.xlsx", sheet_name=("datos1"))
db["id"] = id
db.set_index('id', inplace=True)
db
```

```python
v = db['N']
F1 = db['F1']
F2 = db['F2']

plt.plot(v,F1, color = "c")
plt.plot(v,F2, color = "r")
plt.xlabel("Velocidad")
plt.ylabel("Torque")

plt.title("Gráfico Torque - velocidad")
plt.savefig("Graf1")
plt.plot()
```
