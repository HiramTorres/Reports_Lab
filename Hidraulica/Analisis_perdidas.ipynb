## Muestra para datos de 25% de velocidad  (375 rpm)

# An√°lisis de perdidas 


```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme(style="darkgrid")

id = pd.Series(range(1,9) , name="id")
df25 = pd.read_excel("PRACTICA.xlsx", sheet_name="25%")
df50 = pd.read_excel("PRACTICA.xlsx", sheet_name="50%")
df75 = pd.read_excel("PRACTICA.xlsx", sheet_name="75%")
df100 = pd.read_excel("PRACTICA.xlsx", sheet_name="100%")

df25['id'] = id
df25.set_index('id', inplace=True)
data25 = df25[['Pump Speed n[rpm ]','Flow Rate Q [I/S]','Outlet Pressure Pout [KPa]', 'Outley Velocity Vout [m/s]','System Head Loss[m]']]
data25
```

```python
Q25 = data25['Flow Rate Q [I/S]']
P_out25 = data25['Outlet Pressure Pout [KPa]']
Vel25 = data25['Outley Velocity Vout [m/s]']
HLoss25 = data25['System Head Loss[m]']

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(Q25, P_out25, color='c')
ax[0].set_xlabel('Flow Rate [I/S]')
ax[0].set_ylabel('Pressure out [KPa]')

ax[1].plot(Q25, Vel25, color = 'r')
ax[1].set_xlabel('Flow Rate [I/S]')
ax[1].set_ylabel('Velocity [m/s]')

ax[2].plot(Q25, HLoss25, color = 'b')
ax[2].set_xlabel('Flow Rate [I/S]')
ax[2].set_ylabel('Head Loss [m]')

plt.savefig('Graf25%.png')
plt.show()
```

```python
plt.plot(Q25,P_out25, color = 'c')
plt.plot(Q25,Vel25, color = 'r')
plt.plot(Q25,HLoss25, color = 'b')
plt.xlabel('Flow Rate')
plt.legend(['Pressure', 'Velocity', 'Head Loss'])

plt.savefig('AllGraf25%.png')
plt.show()
```
