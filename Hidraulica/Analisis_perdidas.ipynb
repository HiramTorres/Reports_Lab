# Análisis de perdidas 


## Muestra para datos de 25% de velocidad

```python
import pandas as pd 
import matplotlib.pyplot as plt 

id = pd.Series(range(1,9) , name="id")
df25 = pd.read_excel("PRACTICA.xlsx", sheet_name="25%")
df50 = pd.read_excel("PRACTICA.xlsx", sheet_name="50%")
df75 = pd.read_excel("PRACTICA.xlsx", sheet_name="75%")
df100 = pd.read_excel("PRACTICA.xlsx", sheet_name="100%")

df25['id'] = id
df25.set_index('id', inplace=True)
df25
```

```python
data25 = df25[['Pump Speed n[rpm ]','Flow Rate Q [I/S]','Outlet Pressure Pout [KPa]', 'Outley Velocity Vout [m/s]','System Head Loss[m]']]
data25
```

```python
Q25 = data25['Flow Rate Q [I/S]']
P_out25 = data25['Outlet Pressure Pout [KPa]']
plt.plot(Q25,P_out25, color = 'c')
plt.xlabel('Flow Rate')
plt.ylabel('Outlet Pressure')
plt.title('Pressure vs Flow Rate')
plt.show()
```

## Muestra para datos de 100% de velocidad

```python
df100['id'] = id
df100.set_index('id', inplace=True)
data100 = df100[['Pump Speed n[rpm ]','Outlet Pressure Pout [KPa]', 'Outley Velocity Vout [m/s]','System Head Loss[m]','Flow Rate Q [I/S]']]
data100
```

```python
Q100 = data100['Flow Rate Q [I/S]']
P_out100 = data100['Outlet Pressure Pout [KPa]']
Loss_100 = data100['System Head Loss[m]']

plt.plot(Q100, P_out100, "-",color='c', label='Outlet Pressure')
plt.plot(Q100, Loss_100, "-",color='r', label='System Head Loss')
plt.xlabel('Flow Rate Q [I/S]')
plt.title('Flow Rate  vs Outlet Pressure')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1,2, sharey = True)
ax[0].plot(Q100,P_out100,color = 'c')
ax[0].set_xlabel('Flow Rate Q [I/S]')
ax[0].set_ylabel('Outlet Pressure Pout [KPa]')
ax[0].set_title('Flow Rate  vs Outlet Pressure')
ax[1].plot(Q100,Loss_100,color = 'm')
ax[1].set_xlabel('Flow Rate Q [I/S]')
ax[1].set_ylabel('System Head Loss[m]')
ax[1].set_title('Flow Rate  vs System Head Loss')
plt.show()
```
